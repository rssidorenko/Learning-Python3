```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

heart = pd.read_csv('heart.csv')
heart.head()
```

```python
num_features = ['age', 'trestbps', 'chol', 'thalach']
cat_features = ['sex', 'cp', 'restecg', 'exang', 'target']
```

```python
sns.distplot(heart['age'])
```

```python
heart['target'].value_counts()
```

```python
sns.barplot(x=heart['target'].value_counts().index, y=heart['target'].value_counts())
```

```python
sns.violinplot(x=heart['target'], y=heart['age'])
```

```python
sns.boxplot(x=heart['target'], y=heart['age'])
```

```python
correlation = heart.corr().loc[num_features, num_features]
correlation
```

```python
fig, ax = plt.subplots(figsize = (14,12))

sns.heatmap(correlation, annot=True, linecolor='white', linewidth=0.5, annot_kws = {'size':20},
            vmax=1, vmin=-1)
```

```python
sns.pairplot(heart[num_features], diag_kind='kde')
```

```python
heart_melted = pd.melt(heart, id_vars='age', value_vars=cat_features)
```

```python
def violinplot(x, y, **kwargs):
  sns.violinplot(x=x, y=y)
  x=plt.xticks(rotation=90)

g = sns.FacetGrid(heart_melted, col='variable', col_wrap=2, sharex=False, sharey=False, height=5)

g.map(violinplot, 'value', 'age')
```
