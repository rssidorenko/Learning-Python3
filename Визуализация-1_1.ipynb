```python
def Season(x: int) -> str:

    """
    this function converts number of month to season
    :param i: number of month
    :type i: int
    :return: name of season
    :rtype: str
    """

    if x in [1, 2, 12]:
        return'Winter'
    elif x in range (3, 6):
        return 'Spring'
    elif x in range (6, 9):
        return 'Summer'
    return 'Fall'

Court['Season'] = Court['Month'].apply(Season)
Court
```

```python
Court_new = pd.DataFrame(Court.groupby(['Season', 'year'])['Title'].count())
Court_new
```

```python
Court_new.groupby('Season').mean().plot(kind='bar')
```
