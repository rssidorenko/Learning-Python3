```python
import pandas as pd
X = [13, 21, 30, 65, 50]
Y = [200, 90, 10, 83, 12]
X1 = pd.Series(X)
Y1 = pd.Series(Y)
X1, Y1
```

```python
new_table = pd.concat([X1, Y1], axis=1)
new_table.columns = ['X1', 'Y1']
new_table
```

```python
new_table['X1'].corr(new_table['Y1'])
```

```python
new_table = new_table.rename(columns={'X1': 'Критерий 1', 'Y1': 'Критерий 2'})
new_table
```
