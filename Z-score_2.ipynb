```python
import numpy as np
import pandas as pd
Marvel = pd.read_csv('MarvelComicCaracters.csv', sep=';', comment='#')
Marvel
```

```python
Marvel.isna()['appearances'].sum()
```

```python
Marvel = Marvel.dropna(subset=['appearances'])
q1 = np.percentile(Marvel['appearances'], 25)
q3 = np.percentile(Marvel['appearances'], 75)
iqr = q3 - q1
Marvel.groupby('align')['appearances'].agg(lambda x: iqr)
```

```python
Marvel = Marvel.dropna(subset=['appearances'])
pos = Marvel[Marvel['align'] == 'Добрый персонаж']
q1 = np.percentile(pos['appearances'], 25)
q3 = np.percentile(pos['appearances'], 75)
iqr = q3 - q1
iqr_threshold_bottom = q1 - 1.5*iqr
iqr_threshold_top = q3 + 1.5*iqr
bottom = Marvel[Marvel['appearances'] <= iqr_threshold_bottom].shape
top = Marvel[Marvel['appearances'] >= iqr_threshold_top].shape
print(bottom, top)
#Marvel.groupby('align')['appearances'].agg(lambda x: iqr)
```

```python
neut = Marvel[Marvel['align'] == 'Нейтральный персонаж']
q1 = np.percentile(neut['appearances'], 25)
q3 = np.percentile(neut['appearances'], 75)
iqr = q3 - q1
print(iqr)
iqr_threshold_bottom = q1 - 1.5*iqr
iqr_threshold_top = q3 + 1.5*iqr
bottom = Marvel[Marvel['appearances'] <= iqr_threshold_bottom].shape
top = Marvel[Marvel['appearances'] >= iqr_threshold_top].shape
print(bottom, top)
```

```python
neg = Marvel[Marvel['align'] == 'Злой персонаж']
q1 = np.percentile(neg['appearances'], 25)
q3 = np.percentile(neg['appearances'], 75)
iqr = q3 - q1
print(iqr)
iqr_threshold_bottom = q1 - 1.5*iqr
iqr_threshold_top = q3 + 1.5*iqr
bottom = Marvel[Marvel['appearances'] <= iqr_threshold_bottom].shape
top = Marvel[Marvel['appearances'] >= iqr_threshold_top].shape
print(bottom, top)
```

```python
Q3 = Marvel.groupby('align')['appearances'].quantile(q=0.75)
Q1 = Marvel.groupby('align')['appearances'].quantile(q=0.25)
iqr = Q3 - Q1
print(iqr)
```

```python
Q31 = Marvel['appearances'].quantile(q=0.75)
Q11 = Marvel['appearances'].quantile(q=0.25)
iqr1 = Q31 - Q11
iqr1
```

```python
iqr_threshold_bottom = Q11 - 1.5*iqr1
iqr_threshold_top = Q31 + 1.5*iqr1
bottom = Marvel[Marvel['appearances'] <= iqr_threshold_bottom].shape
top = Marvel[Marvel['appearances'] >= iqr_threshold_top].shape
print(bottom, top)
```

```python
Marvel['app_log'] = np.log(Marvel['appearances'])
Marvel = Marvel.dropna(subset=['app_log'])
std = Marvel['app_log'].std()
mean = Marvel['app_log'].mean()
std, mean
```

```python
std_threshold_bottom = mean - 3 * std
std_threshold_top = mean + 3 * std
print(std_threshold_bottom, std_threshold_top)
```

```python
std_bottom = Marvel[Marvel['app_log'] >= std_threshold_top].shape
std_top = Marvel[Marvel['app_log'] <= std_threshold_bottom].shape
std_bottom, std_top
```

```python
std_1 = Marvel['appearances'].std()
mean_1 = Marvel['appearances'].mean()
std_1, mean_1
```

```python
std_threshold_bottom_1 = mean_1 - 3 * std_1
std_threshold_top_1 = mean_1 + 3 * std_1
print(std_threshold_bottom_1, std_threshold_top_1)
```

```python
std_bottom_1 = Marvel[Marvel['appearances'] >= std_threshold_top_1].shape
std_top_1 = Marvel[Marvel['appearances'] <= std_threshold_bottom_1].shape
std_bottom_1, std_top_1
```

```python
import scipy.stats
Marvel['app_zscore'] = scipy.stats.zscore(Marvel['appearances'])
fem = Marvel[(Marvel['sex'] == 'Женский персонаж') & ((Marvel['app_zscore'] >= 3) | (Marvel['app_zscore'] <= -3))]
fem.count()
```

```python
Q31 = Marvel['appearances'].quantile(q=0.75)
Q11 = Marvel['appearances'].quantile(q=0.25)
iqr1 = Q31 - Q11
threshold_bottom = Q11 - 1.5*iqr1
threshold_top = Q31 + 1.5*iqr1
Marvel_new = Marvel[(Marvel['appearances'] <= threshold_top) & (Marvel['appearances'] >= threshold_bottom)]
Marvel_new['appearances'].mean()
```
