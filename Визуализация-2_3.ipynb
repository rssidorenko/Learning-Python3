```python
import pandas as pd
urban = pd.read_csv('UrbanizationCensusTract.csv')
urban.head()
```

```python
urban_Alaska = urban[urban['state'] == 'Alaska']
urban_Alaska
```

```python
import seaborn as sns
```

```python
urban_new1 = urban[(urban['state'] == 'California') | (urban['state'] == 'Nebraska') 
| (urban['state'] == 'Arizona') | (urban['state'] == 'Texas')]
urban_new1.head()
```

```python
sns.boxplot(x=urban_new1['state'], y=urban_new1['urbanindex'])
```

```python
new_features = ['population', 'adj_radiuspop_5', 'urbanindex']
correlation = urban.corr().loc[new_features, new_features]
correlation
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (10, 8))

sns.heatmap(correlation, annot=True, linecolor='white', linewidth=0.5, annot_kws = {'size':20},
            vmax=1, vmin=-1)
```

```python
import scipy.stats
```

```python
urban_new3['zscore'] = scipy.stats.zscore(urban_new3['urbanindex'])
urban_new3.head()
```

```python
urban_new3z = urban_new3[(urban_new3['zscore'] >= 3) | (urban_new3['zscore'] <= -3)]
urban_new3z.value_counts()
```
